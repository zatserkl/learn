#### .loc usage 
This is a really powerful and flexible method

```python
# .loc DataFrame method
# filtering rows and selecting columns by label

# format
# ufo.loc[rows, columns]

# row 0, all columns
row1 = ufo.loc[0, :]
print('type(row1):', type(row1))
print(row1)
```

```python
# ufo.loc[ufo["Colors Reported"] == "GREEN"]
# ufo.loc[ufo["Colors Reported"].isnull(), :]   # select rows with 'Color Reported' is NaN
ufo.loc[ufo['Colors Reported'].notnull(), :]  # select rows with 'Color Reported' is not NaN
# ufo.dropna(thresh=2)  # threshold = 2 NaNs
```

```python
# filter using City=='Oakland'
ufo[ufo.City=='Oakland']
```

```python
# easier-to-read code
# here you specify the rows and columns you want
# ufo.loc[rows, columns]

ufo.loc[ufo.City=='Oakland', :]
```

```python
titles = ['Movie A', 'Movie B', 'Movie C']
genres = ['Comedy', 'Triller, Horror', 'Drama, Comedy']
movies = pd.DataFrame({'Title': titles, 'Genre': genres})  # create DataFrame
print(movies)
print()

for genre in movies['Genre']:
    print('genre:  ', genre)
    for genre_str in str(genre).split(','):  # genre_str is an element of the list str(genre).split(',')
        genre_str = genre_str.strip()
        # print(genre_str)
        movies[genre_str] = movies['Genre'].str.contains(genre_str)  # solution with True, False

print(movies)
print(movies.columns)
```

```python
# use np.where

import numpy as np

titles = ['Movie A', 'Movie B', 'Movie C']
genres = ['Comedy', 'Triller, Horror', 'Drama, Comedy']
movies = pd.DataFrame({'Title': titles, 'Genre': genres})  # create DataFrame
print(movies)
print()

for genre in movies['Genre']:  # type(movies["Genre"]): <class 'pandas.core.series.Series'>
    print('genre:  ', genre, '   type(genre):', type(genre))
    for genre_str in genre.split(','):
        genre_str = genre_str.strip()
        movies[genre_str] = np.where(movies['Genre'].str.contains(genre_str), 1, 0)  # solution with 1, 0

print(movies)
```

```python
# List comprehension solution, Answer #59 from https://stackoverflow.com/questions/19913659/pandas-conditional-creation-of-a-series-dataframe-column

import numpy as np

titles = ['Movie A', 'Movie B', 'Movie C']
genres = ['Comedy', 'Triller, Horror', 'Drama, Comedy']
movies = pd.DataFrame({'Title': titles, 'Genre': genres})  # create DataFrame
print(movies)
print()

for genre in movies['Genre']:  # type(movies["Genre"]): <class 'pandas.core.series.Series'>
    print('genre:  ', genre, 'type(genre):', type(genre))
    for genre_str in genre.split(','):
        genre_str = genre_str.strip()
        movies[genre_str] = [1 if x.find(genre_str) >= 0 else 0 for x in movies['Genre']]   # solution with 1, 0

print(movies)
```

```python
# map lambda function, Answer #59, 60 from https://stackoverflow.com/questions/19913659/pandas-conditional-creation-of-a-series-dataframe-column

import numpy as np

titles = ['Movie A', 'Movie B', 'Movie C']
genres = ['Comedy', 'Triller, Horror', 'Drama, Comedy']
movies = pd.DataFrame({'Title': titles, 'Genre': genres})  # create DataFrame
print(movies)
print()

for genre in movies['Genre']:  # type(movies["Genre"]): <class 'pandas.core.series.Series'>
    print('genre:  ', genre)
    for genre_str in genre.split(','):
        genre_str = genre_str.strip()
        movies[genre_str] = movies['Genre'].map(lambda x: 1 if x.find(genre_str) > 0 else 0)  # solution with 1, 0

print(movies)
```

```python
# Complete Answer #59 from https://stackoverflow.com/questions/19913659/pandas-conditional-creation-of-a-series-dataframe-column
    
import pandas as pd
import numpy as np

df = pd.DataFrame({'Type':list('ABBC'), 'Set':list('ZZXY')})
print(df)
%timeit df['color'] = ['red' if x == 'Z' else 'green' for x in df['Set']]    # list comprehension
%timeit df['color'] = np.where(df['Set']=='Z', 'green', 'red')               # np.where
%timeit df['color'] = df.Set.map( lambda x: 'red' if x == 'Z' else 'green')  # map lambda function
print(df)
```

### Cookbook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

df = pd.DataFrame({'AAA': [4,5,6,7], 'BBB': [10, 20, 30, 40], 'CCC': [100, 50, -30, -50]})
display(df)
```

```python
"""NB below: the condition df.AAA > 5 and the column to change are separated by comma
"""
df.loc[df.AAA >= 5, 'BBB'] = -1  # conditional change of the column 'BBB': col name 'BBB' should be the last
display(df)
```

```python
df.loc[df.AAA >= 5, ['BBB', 'CCC']] = 555  # conditional change of the columns from list ['BBB', 'CCC']
display(df)
```

```python
df.loc[df.AAA < 5,['BBB','CCC']] = 2000; df
```

```python
df = pd.DataFrame({'AAA': [4,5,6,7], 'BBB': [10, 20, 30, 40], 'CCC': [100, 50, -30, -50]})
display(df)
df_low = df[df.AAA <= 5]  # low part of df
display(df_low)
```
